```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('Digite os parâmetros do problema')

m = float(input('Massa do saltador em kg: '))
t = int(input('Digite o tempo final em segundos: '))
t2 = int(input('Digite a quantidade de intervalos: '))
c = float(input('Coeficiente de arrasto: '))

g = 9.81
```

```python
v_tempo_analitico = np.linspace(0, t, t2)
tam = len(v_tempo_analitico)
v_analitico = np.zeros(tam)

for i in range(tam):
	v_analitico[i] = (np.sqrt((g*m) / c)) * (np.tanh(np.sqrt((g*c) / m) * (v_tempo_analitico[i])))
	print(f'Tempo: {v_tempo_analitico[i]}, Velocidade {v_analitico[i]}')

plt.plot(v_tempo_analitico, v_analitico)
plt.title('Velocidade do Saltador de Bumgee Jumping')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.plot()
```

```python
v_tempo_numerico = np.linspace(0, t, t2)
tam = len(v_tempo_numerico)

v_numerico = np.zeros(tam)
print('Tempo:\t\tVelocidade:')

for i in range(1, tam):
	v_numerico[i] = v_numerico[i-1] + (g - (c/m) * v_numerico[i-1]**2 ) * (v_tempo_numerico[i] - v_tempo_numerico[i-1])

	v_tempo_aux = round(v_tempo_numerico[i], 4)
	v_aux = round(v_numerico[i], 4)
	print(f'{v_tempo_aux}\t\t{v_aux}')

plt.plot(v_tempo_numerico, v_numerico)
plt.title('Velocidade do Saltador de Bumgee Jumping')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.show()
```
