```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

m: int = 8
e_verdadeiro: float = 1.648721
e_tolerado: float = (1 / 2) * math.pow(10, (2-m))
e_05: list[float] = [0.0]
abscissa: list[int] = [0]
taxa: float = 0.5
```

```python
indice: int = 0
while (np.abs(e_verdadeiro - e_05[indice]) >= e_tolerado):
	e_05.append(e_05[indice] + float(np.pow(taxa, indice) / math.factorial(indice)))
	indice += 1
	abscissa.append(indice)

print(f"Valor\t\t\tPassos")
for i in range(0, len(e_05)):
	print(f"{e_05[i]:.8f}\t\t{abscissa[i]}")

plt.plot(abscissa, e_05)
plt.title('Valor de e^0.5')
plt.xlabel('Passos')
plt.ylabel('Valor estimado')
plt.plot()
```
