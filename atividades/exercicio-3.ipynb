```python
import matplotlib.pyplot as plt
```

```python
passos: float = 1.0
x_atual: float = 1.0
diferenca: list[float] = [0.0]
abssissa: list[int] = [0]

def funcao(x: float) -> float:
	return (-0.1 * (x ** 4)) - (0.15 * (x ** 3)) - (0.5 * (x ** 2)) - (0.25 * x) + 1.2

def derivada(x: float) -> float:
	return (-0.4 * (x ** 3)) - (0.45 * (x ** 2)) - (1.0 * x) - 0.25
```

```python
print("Função\t\tDerivada Calculada\t\tDiferenca Centrada")
for i in range(0, 20):
	abssissa.append(i)
	diferenca.append((funcao(i+1) - funcao(i-1)) / (2.0 * passos))
	print(f"{funcao(i):.6f}\t\t{derivada(i):.6f}\t\t\t{diferenca[i]:.6f}")

plt.plot(abssissa, diferenca)
plt.title("Aproximação de f'(x)")
plt.xlabel("Passos")
plt.ylabel("Valor Aproximado")
plt.show()
```
